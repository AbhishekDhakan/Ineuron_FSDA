Q2. What are the different ways you can select columns and rows ?

Ans ->

There are several ways to select columns and rows in each catering to different needs and preferences.

let us go through all ways of it as given below :-

1. **Click and Drag**: 
   - By clicking on the header of the column letter or row number and dragging across adjacent columns or rows, you can select them.
   - This method is straightforward and intuitive, making it easy to quickly select contiguous columns or rows.

2. **Ctrl + Click**:
   - By holding down the Ctrl key on your keyboard and clicking on the column letters or row numbers, you can select multiple non-contiguous columns or rows.
   - This method allows you to select specific columns or rows that are spread throughout the worksheet.

3. **Shift + Click**:
   - By holding down the Shift key on your keyboard and clicking on the header of the column letter or row number, you can select a contiguous range of columns or rows.
   - This method is handy for selecting a large block of adjacent columns or rows all at once.

4. **Ctrl + Shift + Arrow Keys**:
   - By holding down the Ctrl and Shift keys simultaneously and then pressing the arrow keys (up, down, left, or right), you can extend the selection to adjacent columns or rows.
   - This method offers precise selection control, particularly when navigating to specific areas of the worksheet.

5. **Select All**:
   - By clicking on the triangle at the intersection of the row numbers and column letters (top-left corner of the worksheet), you can select the entire worksheet.
   - This method is useful when you need to select all columns and rows in the worksheet, such as when applying formatting or performing operations on the entire dataset.

6. **Keyboard Shortcuts**:
   - Keyboard shortcuts like Ctrl + Spacebar and Shift + Spacebar can be used to select the entire column or row of the active cell, respectively.
   - These shortcuts offer quick and easy access to selecting entire columns or rows witho and workflow preferences.

Q3. What is AutoFit and why do we use it ?

Ans ->

AutoFit is a functionality that automatically modifies the width of columns or the height of rows to accommodate the contents within them. It guarantees that all content within a cell is displayed without being cut off or clipped, enhancing the legibility and appearance of your spreadsheet. This feature eases the formatting process by conserving time and effort that would otherwise be spent on manually adjusting each column or row one by one. Regardless of whether you're working with text, numbers, or other data types, AutoFit optimizes your workflow, enabling you to concentrate on other tasks and boosting overall productivity.

Q5. How do you hide and unhide columns in excel ?

Ans ->

This are the ways which leads us to the process that lets you manage the visibility of specific columns within your worksheet. Here's how you can achieve it:

To Conceal Columns:
1. Choose the columns you wish to conceal. This can be done by clicking on the column letters at the top of the worksheet. If you wish to conceal multiple columns, click and drag across the column letters to select them.
2. Right-click on one of the chosen column letters. This will bring up a context menu.
3. From the context menu, select the "Hide" option. Excel will conceal the chosen columns, removing them from sight without erasing any data.

To Reveal Columns:
1. If you wish to reveal a single column, first, choose the columns adjacent to the concealed column. For instance, if you wish to reveal column C, choose columns B and D.
2. Right-click on one of the chosen column letters. This will bring up a context menu.
3. From the context menu, select the "Unhide" option. Excel will display the concealed column.

Alternatively, if you have multiple concealed columns and wish to reveal all of them:
1. Choose the columns surrounding the concealed columns. For instance, if you have concealed columns C, D, and E, choose columns B and F.
2. Right-click on one of the chosen column letters and select "Unhide". Excel will reveal all concealed columns within theive information.
