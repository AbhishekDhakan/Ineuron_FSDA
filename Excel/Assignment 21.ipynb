Q1. Write a VBA code to enter your name in A1 Cell using Input Box and once you enter the name display a message box that says the name has been entered. ?

Ans ->

A VBA code that asks the user to input their name using an Input Box and then shows a message box confirming that the name has been input:

```vba
Sub EnterName()
    Dim userName As String
    
    ' Ask user to input their name using an Input Box
    userName = InputBox("Please input your name:", "Input Name")
    
    ' Check if user input a name
    If userName <> "" Then
        ' Input the name in cell A1
        Range("A1").Value = userName
        
        ' Show a message box confirming that the name has been input
        MsgBox "Your name '" & userName & "' has been input successfully.", vbInformation, "Name Input"
    Else
        ' Show a message box if user cancels or leaves the input box empty
        MsgBox "No name input.", vbExclamation, "Name Not Input"
    End If
End Sub
```

To use this code, simply copy and paste it into a VBA module in Excel, and then run the "EnterName" subroutine. It will ask you to input your name, and once you input the name and click OK, it will show a message box confirming that the name has been input and put the name in cell A1. If you cancel the input box or leave it empty, it will show a message box indicating that no name has been input.

Q2. What are Userforms? Why are they used? How to fill a list box using for loop ?

Ans ->

Userforms in VBA are custom-made dialog boxes that can be developed to interact with users in Excel applications. They offer a graphical interface for users to enter data, make choices, or execute other actions within an Excel workbook. Userforms are utilized to enhance the user experience by offering a more intuitive and user-friendly method to interact with complex applications or carry out specific tasks. They can encompass various controls such as text boxes, buttons, checkboxes, dropdown lists, and list boxes, enabling users to enter data or make choices in an organized manner.

To populate a list box using a For loop in VBA, you can iterate over a range of data or an array and add each item to the list box programmatically. Here's a basic example:

```vba
Sub PopulateListBox()
    Dim myList As Variant
    Dim i As Integer
    
    ' Define the list of items (for demonstration purposes)
    myList = Array("Item 1", "Item 2", "Item 3", "Item 4", "Item 5")
    
    ' Clear the existing items in the list box (optional)
    ListBox1.Clear
    
    ' Loop through the list of items and add each item to the list box
    For i = LBound(myList) To UBound(myList)
        ListBox1.AddItem myList(i)
    Next i
End Sub
```

In this example, we initially define an array called "myList" containing the items to be added to the list box. Then, we clear any existing items in the list box using the "Clear" method (optional). Next, we use a For loop to iterate over each item in the array and add it to the list box using the "AddItem" method. Ultimately, the list box is populated with the items from the array.

Q3. What is an array? Write a VBA code to enter students and their marks from the below table. 

Ans ->

An array in VBA is a collection of variables of the same type that are assembled under a single identifier. It enables you to store multiple values of the same data type in a single variable, simplifying the management and manipulation of large data sets. Arrays can be one-dimensional, two-dimensional, or multi-dimensional, depending on the complexity of the data.

To input students and their marks from a table into an array using VBA, you can define a two-dimensional array where each row signifies a student and each column represents a different piece of information, such as the student's name and their marks. Here's a basic example:

```vba
Sub InputDataIntoArray()
    Dim studentData(1 To 3, 1 To 2) As String
    Dim i As Integer, j As Integer
    
    ' Input student names and marks into the array
    studentData(1, 1) = "John"
    studentData(1, 2) = "85"
    studentData(2, 1) = "Jane"
    studentData(2, 2) = "92"
    studentData(3, 1) = "Tom"
    studentData(3, 2) = "78"
    
    ' Display the array data in the Immediate window (for demonstration purposes)
    For i = 1 To 3
        For j = 1 To 2
            Debug.Print "Student " & i & ": " & studentData(i, j)
        Next j
    Next i
End Sub
```

In this example, we first define a two-dimensional array called "studentData" with dimensions 1 to 3 for students and 1 to 2 for their name and marks. We then manually input the student names and marks into the array. Finally, we use nested For loops to iterate through the array and display the student data in the Immediate window for demonstration purposes.

Q4. Use the following data to create a pie chart using VBA code. Use Font- ‘Times new Roman’, Size -14, Bold, Title - Piechart’ and you are per to use colours as per your taste.

Ans ->

To generate a pie chart using VBA code with the specified parameters, you can utilize the following script:

```vba
Sub GeneratePieChart()
    Dim ws As Worksheet
    Dim chartObj As ChartObject
    Dim rngData As Range
    Dim rngLabels As Range
    Dim i As Integer
    
    ' Specify the worksheet where the chart will be positioned
    Set ws = ThisWorkbook.Sheets("Sheet1") ' Modify "Sheet1" to your sheet name
    
    ' Define the data range and labels range
    Set rngData = ws.Range("A1:A5") ' Modify the range to your data range
    Set rngLabels = ws.Range("B1:B5") ' Modify the range to your labels range
    
    ' Add a new chart object to the worksheet
    Set chartObj = ws.ChartObjects.Add(Left:=100, Width:=375, Top:=75, Height:=225)
    
    ' Specify the chart type and data source
    With chartObj.Chart
        .SetSourceData Source:=rngData
        .ChartType = xlPie
        
        ' Set chart title and font properties
        .HasTitle = True
        .ChartTitle.Text = "Pie Chart"
        .ChartTitle.Font.Name = "Times New Roman"
        .ChartTitle.Font.Size = 14
        .ChartTitle.Font.Bold = True
        
        ' Format data labels
        For i = 1 To rngLabels.Rows.Count
            .SeriesCollection(1).Points(i).DataLabel.Text = rngLabels.Cells(i, 1).Value
            .SeriesCollection(1).Points(i).DataLabel.Font.Name = "Times New Roman"
            .SeriesCollection(1).Points(i).DataLabel.Font.Size = 14
            .SeriesCollection(1).Points(i).DataLabel.Font.Bold = True
        Next i
        
        ' Apply colors to the data points
        .SeriesCollection(1).Points(1).Interior.Color = RGB(255, 0, 0) ' Red
        .SeriesCollection(1).Points(2).Interior.Color = RGB(0, 255, 0) ' Green
        .SeriesCollection(1).Points(3).Interior.Color = RGB(0, 0, 255) ' Blue
        .SeriesCollection(1).Points(4).Interior.Color = RGB(255, 255, 0) ' Yellow
        .SeriesCollection(1).Points(5).Interior.Color = RGB(255, 0, 255) ' Magenta
        
    End With
End Sub
```

This code will generate a pie chart on the specified worksheet with the given data range and labels range. It sets the title font properties to Times New Roman, size 14, and bold. Additionally, it applies custom colors to each data point in the pie chart. Adjust the data range, labels range, and colors as per your requirements.

Q6. Write step by step procedure to protect your workbook using a password ?

Ans ->

To secure your workbook with a password in Excel, adhere to these step-by-step instructions:

1. **Open Workbook**: Launch the Excel workbook that you wish to secure with a password.

2. **Go to Review Tab**: Select the "Review" tab in the Excel ribbon. This tab houses tools associated with workbook protection and security.

3. **Opt for Protect Workbook**: In the "Changes" group within the "Review" tab, find and select the "Protect Workbook" option. This action will trigger a drop-down menu.

4. **Pick Encrypt with Password**: From the drop-down list, pick the "Encrypt with Password" option. This will cause a dialog box to pop up.

5. **Input Password**: In the dialog box, input the password you desire to use to secure the workbook. Ensure to select a robust password that is hard to guess but easy for you to recall. 

6. **Verify Password**: After inputting the password, Excel will request you to verify the password by inputting it once more. This step aids in ensuring that you have not made any typographical errors while inputting the password.

7. **Press OK**: Once you have inputted and verified the password, press the "OK" button. Excel will then encrypt the workbook with the password you supplied.

8. **Save Workbook**: After securing the workbook with a password, ensure to save your changes by pressing the "Save" button or using the keyboard shortcut Ctrl + S. This action will save the workbook with the password protection activated.

9. **Memorize Password**: It's vital to memorize the password you have assigned for the workbook. If you forget the password, it will be difficult to access or modify the workbooFollowing tuture.

Adhering to these steps will assist you in protecting your Excel workbook with a password, ensuring that only authorized users can access its contents.
