Q2. What are the three different types of Boolean operators ?

Ans ->

The three distinct types of Boolean operators are:

1. **AND:** The AND operator yields True if both operands are True, else it yields False. In Python, it's denoted by the keyword `and`.

2. **OR:** The OR operator yields True if at least one of the operands is True, else it yields False. In Python, it's denoted by the keyword `or`.

3. **NOT:** The NOT operator yields the inverse boolean value of the operand. If the operand is True, NOT yields False; if the operand is False, NOT yields True. In Python, it's denoted by the keyword `not`.

These Boolean operators are frequently employed for making logical comparisons and managing the flow of program execution based on conditions.

Q5. What are the six comparison operators ?

Ans ->

The six comparison operators in Python are:

1. **Equal to (`==`):** Verifies if two values are identical.

2. **Not equal to (`!=`):** Verifies if two values are not identical.

3. **Greater than (`>`):** Verifies if the left operand is larger than the right operand.

4. **Greater than or equal to (`>=`):** Verifies if the left operand is larger than or equal to the right operand.

5. **Less than (`<`):** Verifies if the left operand is smaller than the right operand.

6. **Less than or equal to (`<=`):** Verifies if the left operand is smaller than or equal to the right operand.

These operators are frequently employed for making comparisons between values in Python.

Q6. How do you tell the difference between the equal to and assignment operators ? Describe a condition and when you would use one.

Ans ->

In Python, the operator `==` is utilized for comparison, while the operator `=` is utilized for assigning a value to a variable.

Here's how you can distinguish between the two:

- **Equal to operator (`==`):** It's employed to compare two values to verify if they are equal. It returns `True` if the values are equal and `False` otherwise. For instance:
  ```python
  x = 5
  y = 3
  if x == y:
      print("x and y are equal")
  else:
      print("x and y are not equal")
  ```
  Output: `x and y are not equal`

- **Assignment operator (`=`):** It's employed to assign a value to a variable. It takes the value on the right and assigns it to the variable on the left. For instance:
  ```python
  x = 5
  ```
  In this case, `5` is assigned to the variable `x`. 

**Condition Example:**

Suppose you have two variables `temperature` and `threshold`. You want to verify if the temperature surpasses the threshold value. Here's how you would employ both operators:

```python
temperature = 30
threshold = 25

if temperature > threshold:
    print("Temperature exceeds threshold")
else:
    print("Temperature is within threshold")
```

In this example:
- We're employing the greater than operator `>` to compare `temperature` and `threshold` to see if the temperature surpasses the threshold.
- If the temperature is greater than the threshold, we print that the temperature exceeds the threshold. Otherwise, we print that the temperature is within the threshold.

Here, the equal to operator (`==`) is not employed because we're not comparing for equality but rather for a specific relationship between two values.

Q10. How can you tell the difference between break and continue ?

Ans ->

The distinction between `break` and `continue` is based on their impact on loops:

- **`break`:** Upon encountering in a loop, the `break` statement instantly halts the loop and shifts control to the statement immediately succeeding the loop.

- **`continue`:** Upon encountering in a loop, the `continue` statement bypasses the remainder of the current iteration and leaps to the subsequent iteration of the loop, effectively bypassing any remaining code in the loop's body for that specific iteration.

In simpler terms, `break` halts the entire loop, while `continue` bypasses the current iteration of the loop and proceeds to the next iteration.

Q11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1) ?

Ans ->

In Python, `range(10)`, `range(0, 10)`, and `range(0, 10, 1)` all generate the same sequence of numbers from 0 to 9 when used in a `for` loop. However, they differ in their syntax and functionality:

1. **`range(10)`:** This form specifies that the loop should iterate over a sequence of numbers starting from 0 (the default start value) up to, but not including, 10 (the specified stop value). The step value is implicitly set to 1.

2. **`range(0, 10)`:** This form explicitly sets the start and stop values for the sequence. It starts from 0 (the specified start value) and goes up to, but not including, 10 (the specified stop value). As in the previous case, the step value is implicitly set to 1.

3. **`range(0, 10, 1)`:** This form explicitly sets the start, stop, and step values for the sequence. It starts from 0 (the specified start value), goes up to, but not including, 10 (the specified stop value), and increments by 1 (the specified step value) each time.

In essence, all three expressions produce the same sequence of numbers from 0 to 9, but they provide different levels of specificity regarding the start, stop, and step values of the sequence.

Q12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

```python

    print(num)
```

And here's a comparable program that uses a `while` loop to achieve the same outcome:

```python

    print(num)
    num += 1
```

Both programs will yield the output:
```

2
3
4
5
6
7
8

10
```

They essentially fulfill the same objective of printing numbers from 1 to 10, albeit using different loop structures.
