Q10. Make a French-to-English dictionary called f2e from e2f. Use the items method

To construct a French-to-English dictionary named `f2e` from the existing `e2f` dictionary using the `items()` function, you can traverse the items of the `e2f` dictionary and interchange the keys and values. Here's how you can accomplish it:

